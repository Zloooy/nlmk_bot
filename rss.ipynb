```python
import re
import time
import pandas as pd
from pprint import pprint

from gsheet_handler import GSheetHandler
import importlib
module = importlib.import_module('gsheet_handler')
importlib.reload(module)
ggl = module.GSheetHandler('google_sheet_key.json')
```

```python
rss_df_raw = ggl.extract_data_from_google_sheet('https://docs.google.com/spreadsheets/d/1BA1nioQqc048FFvKXcpP5VqL_73kXNCcSx0m-jhI2MQ/edit#gid=2038294867', 
    0)
rss_df_raw
```

```python
rss_df = rss_df_raw[rss_df_raw['rss'].str.contains('http')].reset_index(drop=True)
rss_df
```

```python
import feedparser
import pandas as pd



def parse_rss(url):
    feed = feedparser.parse(url)
    news_items = []
    for entry in feed.entries:
        news_items.append({
            'title': entry.title,
            'link': entry.link,
            'published': entry.published,
            'source': feed.feed.title,
            'content': entry.summary if 'summary' in entry else ''
        })
    return news_items

# Сбор всех новостей в один список
all_news = []
for _, row in rss_df.iterrows():
    news_items = parse_rss(row['rss'])
    all_news.extend(news_items)

# Преобразование списка новостей в DataFrame
news_df = pd.DataFrame(all_news)

# Вывод таблицы
news_df
```

```python
news_df['content'] = news_df['content'].str[:3000]
```

```python
ggl.update_gsheet_with_df(news_df,
                          'https://docs.google.com/spreadsheets/d/1BA1nioQqc048FFvKXcpP5VqL_73kXNCcSx0m-jhI2MQ/edit#gid=0', 
                        1)
```

```python
from dateutil import parser
# Функция для преобразования дат в формат YYYY-MM-DD
def convert_to_date(date_str):
    try:
        parsed_date = parser.parse(date_str)
        return parsed_date.strftime('%Y-%m-%d')
    except ValueError:
        return None

# Применяем функцию к колонке 'dates'
news_df['formatted_dates'] = news_df['published'].apply(convert_to_date)
```

```python
ggl.update_gsheet_with_df(news_df,
                          'https://docs.google.com/spreadsheets/d/1BA1nioQqc048FFvKXcpP5VqL_73kXNCcSx0m-jhI2MQ/edit#gid=0', 
                        1)
```
