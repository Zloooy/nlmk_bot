```python
rss_df = rss_df_raw[rss_df_raw['rss_feed'].str.contains('http')].reset_index(drop=True)
rss_df
```

```python
import feedparser
import pandas as pd
from tqdm import tqdm


def parse_rss(url):
    feed = feedparser.parse(url)
    news_items = []
    for entry in feed.entries:
        news_items.append({
            'title': entry.title,
            'link': entry.link,
            'published': entry.published,
            'source': feed.feed.title,
            'content': entry.summary if 'summary' in entry else ''
        })
    return news_items

# Сбор всех новостей в один список
all_news = []
for _, row in tqdm(rss_df.iterrows(), total=rss_df.shape[0]):
    news_items = parse_rss(row['rss_feed'])
    all_news.extend(news_items)

# Преобразование списка новостей в DataFrame
news_df = pd.DataFrame(all_news)

# Вывод таблицы
news_df
```

```python
news_df['content'] = news_df['content'].str[:40000]
news_df.to_csv('rss.csv')
```

```python
from dateutil import parser
# Функция для преобразования дат в формат YYYY-MM-DD
def convert_to_date(date_str):
    try:
        parsed_date = parser.parse(date_str)
        return parsed_date.strftime('%Y-%m-%d')
    except ValueError:
        return None

# Применяем функцию к колонке 'dates'
news_df['formatted_dates'] = news_df['published'].apply(convert_to_date)
```

```python
import pandas as pd
import tiktoken

# Function to count tokens
def count_tokens(row):
    encoding = tiktoken.get_encoding("cl100k_base")
    text = ' '.join(map(str, row.values))  # Combine all columns into one text
    tokens = encoding.encode(text)
    return len(tokens)

# Function to calculate cost
def calculate_cost(tokens, model):
    if model == "gpt-4o":
        cost_per_token_input = 5 / 1_000_000  # Cost per token for GPT-4o input
        cost_per_token_output = 15 / 1_000_000  # Cost per token for GPT-4o output
        cost_input = tokens * cost_per_token_input
        cost_output = tokens * cost_per_token_output
        return cost_input + cost_output
    elif model == "gpt-3.5-turbo":
        cost_per_token = 0.002 / 1000  # Cost per token for GPT-3.5-turbo
        return tokens * cost_per_token

news_df['content'] = news_df['content'].str[:1000]
news_df['tokens'] = news_df.apply(count_tokens, axis=1)
news_df['cost_gpt4o'] = news_df['tokens'].apply(lambda x: calculate_cost(x, "gpt-4o"))
news_df['cost_gpt35'] = news_df['tokens'].apply(lambda x: calculate_cost(x, "gpt-3.5-turbo"))

news_df
```
